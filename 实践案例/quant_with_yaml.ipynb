Pipeline是scutquant v0.4.6开始推出的一个高度集成的模块. 旨在让使用者无需了解各模块的具体实现的情况下, 仅填写yaml文件即可使用scutquant完成量化研究的全流程.   

Pipeline中包含一个集成所有功能的函数pipeline. 也有在其它模块基础上初步封装的函数, 例如load_data, get_factors等. 可根据自己的需要灵活选择函数搭建自己的workflow.   

## 具体要求:  

(1) 一个地址, 需要的所有文件都应放在此目录下. 所有结果也将保存在该目录  

(2) 需要all_kwargs.yaml和data文件. data的后缀暂时仅支持csv, xls, xlsx, pkl, pickle, ftr和feather.  

(3) all_kwargs.yaml需要包含以下键: "data", "factor", "process", "fit", "prepare", "backtest"和"report".  

(4) pipeline函数支持自动生成配置文件. 例如当回测任务使用默认配置时, all_kwargs的"backtest"填null即可. 

(5) 为了更好地了解Pipeline模块, 请阅读源代码
