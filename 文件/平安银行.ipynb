```python
import pandas as pd
import scutquant as q

df = pd.read_excel('D:/Desktop/平安银行.xlsx')
print(df.head(5))
```

```python
print(df.info())
```

```python
df.set_index(['time_tick'], inplace=True)
print(df.head(5))
```

```python
for i in range(1, 11):
    df['price_' + str(i)] = df['price'].shift(i)
    df['volume_tick_' + str(i)] = df['volume_tick'].shift(i)

print(df.head(5))
```

```python
df = q.clean(df)
print(df.head(5))
```

```python
df['ret'] = df['price'] / df['price_1'] - 1
df = df.drop(['volume', 'volume_tick', 'bid_price', 'bid_volume', 'ask_price', 'ask_volume', 'open_interest', 'amount_tick'], axis=1)
print(df.head(5))
```

```python
price = df.pop('price')
```

```python
q.Auto_ts_ana(df, 'ret', windows=20, freq='3sec', lags=12)
```

```python
df = q.make_trend(df, order=1)
df = q.make_fourier_features(df, freq=240, order=4)
df = q.make_lags(df, data=df['ret'], lags=1)
print(df.head(5))
```

```python
x_train, x_test, y_train, y_test, ymean, ystd = q.AutoProcessor(df, 'ret', norm='r')
```

```python
model  = q.hybrid()
model.fit(x_train, y_train, x_test, (y_test-ymean)/ystd)
```

```python
pred = model.predict(x_test)
pred = pd.Series(pred, index=x_test.index)
pred += ymean
pred *= ystd
print(pred.head(5))
print(y_test.head(5))
```

```python
print(q.pearson_corr(pred, y_test))
```

```python
ic, icir, rank_ic, rank_icir = q.ic_ana(pred, y_test, freq=60)
print(ic, icir, rank_ic, rank_icir)
```
